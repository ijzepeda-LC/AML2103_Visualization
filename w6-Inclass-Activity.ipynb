#### V.III Grouped Bar graph / Multiple-category bar graph

```python
course_list=['CSE','Mech','Aero','Biomed','EXTC']

boys_cnt_height=[100,123,54,299,26]
girls_cnt_height=[33,100,10,140,300]

width=0.4 #of the bar



bar1= np.arange(len(course_list)) #0,1,2,3,4 # for boys
bar2= [(i+width)for i in bar1]

#where the bars are centered - ?

#I can decide where is each bar is located, left or right

# this technique, will locate by values the bars, first one (bar1) by numbers 0-4, and bar2, 0-4+0.4, so 1.4, 2.4, 
# when it is ploted, then items will be "aligned" in X



#plt x heights[values], widht[of bar]
plt.bar( bar1, boys_cnt_height, width, label="Boys")
plt.bar( bar2, girls_cnt_height, width, label="Girls")


# TO locate the Xtick in the middle, and not in the first element:
# this value + (w/2)
plt.xticks(bar1 + (width/2), labels=course_list)

plt.xlabel('Courses')
plt.ylabel('#strudents')
plt.title("students v/s Courses")
plt.legend()
```
