```python
data2 = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [10, 20, 15, 25, 12, 18, 9, 22]
}

df2 = pd.DataFrame(data2)

# Group by 'Category' and aggregate the values
grouped = df2.groupby('Category')['Value'].sum().reset_index()

# Plotting using seaborn
sns.barplot(data=grouped, x='Category', y='Value')
plt.xlabel('Category')
plt.ylabel('Sum of Values')
plt.title('Sum of Values by Category')
plt.show()
```
