#### I) Line graph: 

1.1) Plotting:
* Syntax: `plt.plot([x], y, [fmt_string])`
* Given `n` = #points; x = (0....n-1)

```python
plt.plot([0,1,2,3],[2,4,6,8],'bs:')
```

```python
plt.plot([2,4,6,8],'bs:')
```

```python
# Line= star markers, red in colour, dashed line.
# X coords = odd nums
# Y coords = even nums
```

```python
plt.plot([1,3,5,7],[2,4,6,8],'r*--')
```

1.2) Plotting MULTIPLE data-pairs:
<br>
`plt.plot([x], y1, [fmt_str1], y2, [fmt_str2] )`


```python
plt.plot([2,4,6,8], 'r*--', [1,5,9,13], 'ms-')
```

1.3) Modifying the appearance of the lines:
* `linewidth`
* `markersize`

```python
plt.plot([2,4,6,8],'bs:')
```

Versus

```python
plt.plot([2,4,6,8],'bs:', linewidth = 5, markersize = 12)
```

1.7) Stockprices data:
<br>
Visualize the trends of each company's closing stock price, over time. 

```python
stock_df = pd.read_csv('U:\\Users\\Reena.Shaw\\Downloads\\stockprices.csv')
```

```python
stock_df.head(3)
```

```python
stock_df.shape
```

```python
stock_df['Name'].unique()
```

```python
google_df = stock_df[stock_df['Name']=='GOOGL']
google_df.shape
```

```python
apple_df = stock_df[stock_df['Name']=='AAPL']
apple_df.shape
```

```python
amazon_df = stock_df[stock_df['Name']=='AMZN']
amazon_df.shape
```

```python
msft_df = stock_df[stock_df['Name']=='MSFT']
msft_df.shape
```

```python
fb_df = stock_df[stock_df['Name']=='FB']
fb_df.shape
```

```python
stock_df['date'].nunique()
```

```python
stock_df['close'].max()
```

```python
plt.figure(figsize = (20,15))
plt.plot('date','close', data = google_df, label = "Google")
plt.plot('date','close', data = apple_df, label = "Apple")
plt.plot('date','close', data = amazon_df, label = "Amazon")
plt.plot('date','close', data = fb_df, label = "Facebook")
plt.plot('date','close', data = msft_df, label = "Microsoft")
plt.legend(fontsize = 20) #loc = "upper right"

plt.xticks(np.arange(0,1260, 80))
plt.yticks(np.arange(0,1470, 70))

plt.title('Closing stock prices of 5 companies', fontsize = 20)
plt.ylabel('Closing price in USD', rotation = 0, labelpad = 60)

plt.grid()

plt.show()
```
