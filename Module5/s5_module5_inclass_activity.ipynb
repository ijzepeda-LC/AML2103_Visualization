### 1.7) Stockprices data:
Visualize the trends of each company's closing stock price over time

```python
# Ugly approach
plt.figure(figsize=(15,5))
plt.plot(df['date'],df['close'])
```

**XTICKS and frequency interval**  
Make sure you have your dates in date format  
then xticks should be populated using np.arange(), and the frequency  
in this case, frequency is 60 days in datetime, so '60D' works  
np.arange() consists of an interval of values, from minumum to maximum.  
use np.min(df['col']) and np.max(df['col']) 

```python

df['date'] = pd.to_datetime(df['date'])

plt.figure(figsize=(15,8))
for i in df['Name'].unique():
    plt.plot(df.loc[df['Name']==i,'date'], df.loc[df['Name']==i, 'close'] , label=i)
plt.xlabel('Date')
plt.ylabel('Close (USD)')

# Set the x-axis tick positions
# method 1
date_ticks = pd.date_range(start=np.min(df['date']), end=np.max(df['date']), freq='60D')
plt.xticks(date_ticks, rotation=45)

#method 2
# plt.xticks(np.arange(0,1260), 60) # 1260=> len(df['date'])


plt.title("Company's closing stock price - Ivan Zepeda")
plt.grid()
plt.legend() #loc=" upper right"
plt.savefig("companys-closing-stock-price.png")
plt.show()
```
